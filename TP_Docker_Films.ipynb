# TP Docker – Films

Ce projet a pour objectif d'apprendre à utiliser **Docker**, **Git** et **Python** pour conteneuriser deux applications :
- une application **write** pour écrire / modifier / supprimer des films ;
- une application **read** pour lire et filtrer les films.

Le tout est orchestré par **Docker Compose**, et les données sont partagées grâce à un **volume Docker**.


## Clonage du repository

Pour récupérer le projet :

```bash
git clone https://github.com/<ton-utilisateur>/docker_tp.git
cd docker_tp


## Construction et lancement des conteneurs

Construire les images :

```bash
docker-compose build

docker-compose up // Pour lancer le conteneur

docker ps // Lister les conteneurs actif
docker-compose down // Arrêter et supprimer les conteneurs 

## Application WRITE – Manipulation des données

Le script `manip_data.py` permet d’ajouter, modifier et supprimer des films dans le fichier `movies.csv`.

### Commande Docker
```bash
docker exec -it write_app python manip_data.py

## Application READ – Lecture des données

Le script `read_data.py` permet de filtrer et lire les films selon différents critères.

### Commande Docker
```bash
docker exec -it read_app python read_data.py

## Volume Docker

Les deux conteneurs partagent un même volume nommé `data_volume`.

Ainsi, lorsqu’un film est ajouté depuis `write`, il devient immédiatement visible depuis `read`.
